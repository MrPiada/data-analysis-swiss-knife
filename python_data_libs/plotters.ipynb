## DATA VISUALIZATION: tips and tricks

### Matplotlib: mosaic vs subplots

```python
import matplotlib.pyplot as plt
import numpy as np

# Dati di esempio
x = np.linspace(0, 10, 100)
y = np.sin(x)
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 30, 40]

# Creazione della figura
fig = plt.figure(figsize=(10, 8))

# Scatter plot (AA)
ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=2)
ax1.scatter(x, y)
ax1.set_title('Scatter plot')

# Line plot (E)
ax2 = plt.subplot2grid((3, 3), (0, 2), rowspan=3)
ax2.plot(x, y)
ax2.set_title('Line plot')

# Bar plot (C)
ax3 = plt.subplot2grid((3, 3), (1, 0))
ax3.bar(categories, values)
ax3.set_title('Bar plot')

# Pie plot (.)
ax4 = plt.subplot2grid((3, 3), (1, 1))
ax4.pie(values, labels=categories, autopct='%1.1f%%')
ax4.set_title('Pie plot')

# Regolazione degli spazi tra i subplot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dati di esempio
x = np.linspace(0, 10, 100)
y = np.sin(x)
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 30, 40]

# Definizione del layout dei subplot con una stringa
layout = """
AAE
COE
..E
"""

# Creazione della figura con subplot disposti secondo il layout
fig, ax_dict = plt.subplot_mosaic(layout, figsize=(10, 8))

# Scatter plot (AA)
ax_dict['A'].scatter(x, y)
ax_dict['A'].set_title('Scatter plot')

# Line plot (E)
ax_dict['E'].plot(x, y)
ax_dict['E'].set_title('Line plot')

# Bar plot (C)
ax_dict['C'].bar(categories, values)
ax_dict['C'].set_title('Bar plot')

# Pie plot (O)
ax_dict['O'].pie(values, labels=categories, autopct='%1.1f%%')
ax_dict['O'].set_title('Pie plot')

# Regolazione degli spazi tra i subplot
plt.tight_layout()
plt.show()
```

### Seaborn: kde and pairplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generiamo i dati
np.random.seed(42)  # Imposta un seed per la riproducibilit

# Dati distribuiti normalmente
data_normal = np.random.normal(loc=0, scale=1, size=10)

# Dati distribuiti uniformemente
data_uniform = np.random.uniform(low=-3, high=3, size=50)

# Dati distribuiti esponenzialmente
data_exponential = np.random.exponential(scale=1, size=50)

# Uniamo i dati in un unico array e creiamo una colonna per il tipo di distribuzione
data = np.concatenate([data_normal, data_uniform, data_exponential])

# Creiamo il grafico KDE
sns.displot(data=data, fill=True, label='Normale', rug=True, color='blue', alpha=0.3, kind='kde')


# Aggiungiamo titoli e etichette
plt.title('Grafico KDE e Distribuzioni dei Dati')
plt.xlabel('Valori')
plt.ylabel('Densit')
plt.legend(title='Distribuzione')

# Mostriamo il grafico
plt.show()
```

```python
iris = sns.load_dataset("iris")
g = sns.pairplot(iris, hue="species", palette="Set2", diag_kind="kde", height=2.5)
```

### Interactive plots with Plotly

```python
import plotly.express as px

df = px.data.gapminder().query("continent=='Oceania'")

fig = px.line(df, x="year", y="lifeExp", color="country", title="layout.hovermode='x unified'")
fig.update_traces(mode="markers+lines", hovertemplate=None)
fig.update_layout(hovermode="x unified")

fig.show()
```
