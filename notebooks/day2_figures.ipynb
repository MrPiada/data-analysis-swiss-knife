```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Definiamo il range di t
t = np.linspace(0, 5, 500)

# Definiamo le due distribuzioni (supponiamo che siano gaussiane per semplicità)
g_H0 = norm.pdf(t, loc=1.5, scale=0.5)
g_H1 = norm.pdf(t, loc=3.5, scale=0.8)

# Definiamo il valore di t_cut
t_cut = 2.7

# Creiamo di nuovo la figura con le curve di forme diverse
plt.figure(figsize=(8, 6))

# Tracciamo le distribuzioni con forme diverse
plt.plot(t, g_H0, label=r'$g(t|H_0)$', color='black', linestyle='-', linewidth=2)
plt.plot(t, g_H1, label=r'$g(t|H_1)$', color='black', linestyle='--', linewidth=2)

# Aggiungiamo la linea verticale per t_cut
plt.axvline(t_cut, color='black', linestyle=':', label=r'$t_{cut}$')

# Coloriamo l'area per l'errore di tipo I (alpha)
plt.fill_between(t, 0, g_H0, where=(t >= t_cut), color='red', alpha=0.3, label=r'Tipo I: $\alpha$')

# Coloriamo l'area per l'errore di tipo II (beta)
plt.fill_between(t, 0, g_H1, where=(t < t_cut), color='blue', alpha=0.3, label=r'Tipo II: $\beta$')

# Aggiungiamo le annotazioni per le regioni di accettazione e rifiuto
plt.text(0.5, 1.5, r'accept $H_0$', fontsize=14)
plt.text(3.5, 1.5, r'reject $H_0$', fontsize=14)

# Annotazioni sopra ai picchi delle curve
plt.annotate(r'$g(t|H_0)$', xy=(1.5, max(g_H0)), xytext=(1.8, 1.05),
             arrowprops=dict(facecolor='black', shrink=0.05), fontsize=12)
plt.annotate(r'$g(t|H_1)$', xy=(3.5, max(g_H1)), xytext=(3.8, 0.85),
             arrowprops=dict(facecolor='black', shrink=0.05), fontsize=12)

# Aggiungiamo le etichette degli assi
plt.xlabel(r'$t$', fontsize=14)
plt.ylabel(r'$g(t)$', fontsize=14)

# Aggiungiamo la legenda
l4 = plt.legend(bbox_to_anchor=(0, 1.02, 1, 0.2), loc="lower left",  borderaxespad=0, ncol=5)

# Impostiamo i limiti degli assi
plt.xlim(0, 5)
plt.ylim(0, 2)

# Mostriamo la figura
plt.show()

```
