## Doping test

```python
# Given probabilities
P_D = 1 / 50         # Probability that an athlete uses doping
P_notD = 1 - P_D     # Probability that an athlete does not use doping
P_T_given_D = 0.95   # Probability of testing positive given doping (True Positive)
P_T_given_notD = 0.05  # Probability of testing positive given no doping (False Positive)

# Total probability of testing positive (P(T))
P_T = (P_T_given_D * P_D) + (P_T_given_notD * P_notD)

# Probability of doping given a positive test (P(D|T))
P_D_given_T = (P_T_given_D * P_D) / P_T

P_D_given_T
```

### Doping test confusion matrix

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Hypothetical population size
population = 10000

# True Positives (TP)
TP = P_T_given_D * P_D * population

# False Positives (FP)
FP = P_T_given_notD * P_notD * population

# True Negatives (TN)
TN = (1 - P_T_given_notD) * P_notD * population

# False Negatives (FN)
FN = (1 - P_T_given_D) * P_D * population

# Confusion Matrix
confusion_matrix = {
    "True Positives (TP)": TP,
    "False Positives (FP)": FP,
    "True Negatives (TN)": TN,
    "False Negatives (FN)": FN
}

confusion_matrix


# Create confusion matrix data
confusion_matrix_data = np.array([[TN, FP], [FN, TP]])

# Define labels
labels = ['Predicted Negative', 'Predicted Positive']
categories = ['Actual Negative', 'Actual Positive']

# Plot the confusion matrix
plt.figure(figsize=(10, 8))
sns.heatmap(confusion_matrix_data, annot=True, fmt='g', cmap='Blues', 
            xticklabels=labels, yticklabels=categories, annot_kws={"size": 30})
plt.title('Confusion Matrix', fontsize=18)
plt.xlabel('Predicted', fontsize=20, fontweight='bold')
plt.ylabel('Actual', fontsize=20, fontweight='bold')
plt.xticks(fontsize=14, fontweight='bold')
plt.yticks(fontsize=14, fontweight='bold')
plt.show()
```
