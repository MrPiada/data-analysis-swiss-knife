```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Parametri della distribuzione Gamma
shape, scale = 2., 2.

# Generazione di campioni dalla distribuzione Gamma
np.random.seed(0)
data = np.random.gamma(shape, scale, 1000)

# Calcolo delle statistiche
mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data)[0][0]
variance = np.var(data)
skewness = stats.skew(data)
kurtosis = stats.kurtosis(data)

# Intervallo di varianza
v = np.sqrt(variance)
variance_interval = [mean - v, mean + v]

# Plot della distribuzione e dell'intervallo di varianza
plt.figure(figsize=(16, 6))
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='data')

# Linee verticali per le statistiche
plt.axvline(mean, color='blue', linestyle='--', label=f'Mean ($\\mu$): {mean:.2f}', lw=5)
plt.axvline(median, color='red', linestyle='-.', label=f'Median: {median:.2f}', lw=5)

# Area di varianza
plt.fill_betweenx([0, max(stats.gaussian_kde(data).evaluate(data))], variance_interval[0], variance_interval[1],
                  color='blue', alpha=0.2, label=f'Standard deviation = $\\sqrt{{variance}}$: {v:.2f}')

# Titolo e etichette
plt.title('Example of gamma distribution')
plt.xlabel('X')
plt.ylabel('density')
plt.legend(fontsize=20)
plt.grid(False)
plt.show()

# Output delle statistiche
mean, median, mode, variance, skewness, kurtosis
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, t
from scipy.stats import kurtosis

def plot_continuous_distributions():
    # Define the range for the x-axis
    x = np.linspace(-5, 5, 1000)
    
    # Generate the PDFs for different distributions
    mesokurtic_pdf = norm.pdf(x, loc=0, scale=1)       # Normal distribution: Mesokurtic
    leptokurtic_pdf = t.pdf(x, df=2)                    # T-distribution with df=2: Leptokurtic

    # Calculate kurtosis
    mesokurtic_kurtosis = kurtosis(norm.rvs(size=10000))
    leptokurtic_kurtosis = kurtosis(t.rvs(df=2, size=10000))
    
    plt.figure(figsize=(12, 8))

    # Plot PDFs
    plt.plot(x, mesokurtic_pdf, color='green', label=f'Mesokurtic (Normal)\nKurtosis: {mesokurtic_kurtosis:.2f}', linewidth=8)
    plt.plot(x, leptokurtic_pdf, color='red', label=f'Leptokurtic (t-distribution)\nKurtosis: {leptokurtic_kurtosis:.2f}', linewidth=8)
    
    # Plot formatting
    plt.title('Continuous PDFs of Different Kurtosis Distributions', fontsize=20)
    plt.xlabel('Value')
    plt.ylabel('Density')
    
    # Set legend with a larger font size
    plt.legend(loc='upper right', fontsize=14)
    
    plt.grid(True)
    plt.show()

# Plot the continuous distributions
plot_continuous_distributions()
```
