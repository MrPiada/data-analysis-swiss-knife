```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

x = np.array([1, 2, 3, 4])
y = np.array([2, 5, 10, 20])

def exponential_curve(x, a, b):
    return a * np.exp(b * x)

popt, pcov = curve_fit(exponential_curve, x, y)

x_curve = np.linspace(0, 5, 100)
y_curve = exponential_curve(x_curve, *popt)

with plt.xkcd():
    fig = plt.figure(figsize=(8, 6))



    plt.scatter(x, y, color='blue', label='Punti')
    plt.plot(x_curve, y_curve, 'r--', label='Curva esponenziale')

    for i in range(len(x)):
        plt.annotate(f'({x[i]}, {y[i]})', (x[i], y[i]), textcoords="offset points", xytext=(-10,10), ha='center')

    plt.xlabel('X')
    plt.ylabel('Y')

    plt.style.use('fivethirtyeight')

    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)
    plt.grid(False)
    plt.tight_layout()
    plt.show()
```
