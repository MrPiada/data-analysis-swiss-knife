```python
import matplotlib.pyplot as plt
import numpy as np

# Definizione dei valori x e y
x_values = [1e0, 1e3, 1e6, 1e9, 1e12, 1e15 ]  # KB, MB, GB, TB, PB
y_values = np.exp(np.arange(len(x_values)))  # Crescita esponenziale per la y

# Colori per le aree rettangolari dal verde al rosso
colors = ['#00FF00', '#7FFF00', '#FFFF00', '#FF7F00', '#FF0000']

# Creazione dello scatter plot con lo stile xkcd
with plt.xkcd():
    plt.figure(figsize=(20, 9))
    plt.rcParams.update({'axes.labelsize': 50, 'xtick.labelsize': 40, 'ytick.labelsize': 25})
    
    plt.plot(x_values, y_values, color='blue', linestyle='--', linewidth=10)

    # Impostazione della scala logaritmica per l'asse x
    plt.xscale('log')

    # Rimuove i valori numerici dall'asse y
    plt.yticks([])

    # Etichette degli assi
    plt.xlabel('Data size')
    plt.ylabel('Complexity')

    x_ticks_labels = ['B', 'KB', 'MB', 'GB', 'TB', 'PB']
    plt.xticks(x_values, x_ticks_labels)

    # Aggiunta delle aree rettangolari semi trasparenti
    ranges = [(0, 10), (10, 50), (50, 5000), (100, 2*1e9), (1e6, 1e16)]
    for i, (start, end) in enumerate(ranges):
        plt.axvspan(start, end, color=colors[i], alpha=0.3)

    plt.xlim(0, 1e15+100)
    plt.show()
```
